```python
import itertools
import numpy as np
```

```python
def gen_freq_times(n):
    freqtimes = np.arange(n*2).reshape(n, 2)
    return freqtimes
```

```python
freqtimes = [tuple(i) for i in gen_freq_times(10000).tolist()]
freqtimes
```

```python
combinations = itertools.combinations(freqtimes, 2)
threshold = 2
filtered = itertools.filterfalse(lambda x: x[1][1] - x[0][1] > threshold, combinations)
```
